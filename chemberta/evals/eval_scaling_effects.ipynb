```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
```

```python
df = pd.read_csv("mtr_results_all_seeds.csv")
df["Dataset Size"] = df["pretraining_task"].apply(lambda x: x.split("-")[0])
reorder = dict(zip(["run_45", "run_0", "run_1", "run_38", "run_39"], ["run "+str(x) for x in range(5)]))
df["order"] = df["run_name"].apply(lambda x: reorder[x])

sns.lineplot(data=df, x="order", y="min_eval_loss", hue="Dataset Size", palette="husl", hue_order=["5M", "10M", "77M"], marker="o")
plt.xlabel("Pretraining models sorted by 5M loss")
plt.ylabel("Pretraining Validation Loss")
plt.title("Scaling MTR Pretraining")
plt.show()
```
